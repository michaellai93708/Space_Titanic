```python
import pandas as pd
from statistics import mean
data_train = pd.read_csv('/Users/michael/Desktop/spaceship-titanic/train.csv')
data_test = pd.read_csv('/Users/michael/Desktop/spaceship-titanic/test.csv')

#print(data_train)
#print(data_test)
x = data_train
#, y = data_train.iloc[:, :-1], data_train.iloc[:, [-1]]
x_test = data_test
print(x, y)
print(x_test)
```

```python
group = []
for i in x['PassengerId']:
    pax_group = i.split("_")
    group.append(pax_group[0])
x['Group'] = group

deck, num, side = [], [], []
for i in x['Cabin']:
    if "/" in str(i) :
        pax_cabin = i.split("/")
        deck.append(pax_cabin[0])
        num.append(pax_cabin[1])
        side.append(pax_cabin[2])
    else:
        deck.append(None)
        num.append(None)
        side.append(None)
x["Deck"], x["Num"], x["Side"] = deck, num, side

last_name = []
for i in x["Name"]:
    if " " in str(i):
        pax_name = i.split(" ")
        last_name.append(pax_name[1])
    else:
        last_name.append(None)
x["LastName"] = last_name

x = x.drop(columns=['PassengerId', 'Cabin', 'Name'])
```

```python
x.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encode_list = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'Deck', 'Group',  
               'Num', 'Side', 'LastName']
for i in encode_list:
    x[i] = LabelEncoder().fit_transform(x[i])
x.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer_cols = ["Age", "FoodCourt", "ShoppingMall", "Spa", "VRDeck" ,"RoomService"]
imputer = SimpleImputer(strategy='mean')
imputer.fit(x[imputer_cols])
x[imputer_cols] = imputer.transform(x[imputer_cols])
x.isnull().sum()
```
